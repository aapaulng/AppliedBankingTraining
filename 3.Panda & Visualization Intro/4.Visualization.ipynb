```python
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
```

```python
print('matplotlib version', matplotlib.__version__)
print('seaboarn version', sns.__version__)
```

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.title('Basic Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### Pyplot Interface (Stateful)

```python
plt.plot([1,2,3],[4,5,6])
plt.title('Pyplot Style')
plt.show()
```

### Object Oriented Interface

```python
fig,ax = plt.subplots()
ax.plot([1,2,3],[4,5,6])
ax.set_title('OO Style')
plt.show()
```

### Example Plots in Matplotlib

```python
# Line PLot (Trend)

x = np.linspace(0,10,100)
y = np.sin(x)

# Plot line
plt.plot(x,y,color='blue',linestyle='-',linewidth=2,label='sin(x)')
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Scatter plot (Relationship between variables)

# Generate 50 random points
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y,alpha=0.7,color='blue',marker='o',label='Data Points')
plt.title('Simple Scatter PLot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Histogram (distribution of data)

# Generate 1000 random numbers from normal distribution
data = np.random.randn(1000)

# Plot histogram with 30 bins
plt.hist(data,bins=30,alpha=0.7, density=True, color='Purple', edgecolor='black')
plt.title("Histogram with 30 bins")
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# Seaborn Basics

```python
# Scatter Plot
tips = sns.load_dataset('tips')

# Create scatter plot with color & style differentiation
sns.scatterplot(data=tips,x='total_bill',y='tip', hue='sex', style='time')
plt.title('Scatter Plot of Tips Dataset')
plt.show()
```

```python
# Heatmap (correlation matrix)
iris = sns.load_dataset('iris')
corr = iris.corr(numeric_only=True)

# Draw heatmap with annotatios for clarity
sns.heatmap(corr,annot=True,cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

### Seaborn Styles & Themes

```python
sns.set_style('whitegrid') # Background style
sns.set_palette('muted') # Color palette

sns.barplot(data=tips, x='day',y='total_bill',hue='sex',errorbar=None)
plt.title('Average Bill per Day by gender')
plt.show()
```

```python
# Example of mixing seaborn & matplotlib
ax = sns.histplot(data=tips, x='total_bill', bins=20)
ax.set_title('customized histogram')
plt.show()
```
