```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',9999)
pd.set_option('display.max_rows',9999)
```

```python
df_raw = pd.read_csv('../datasets/credit_score_classification/train.csv')
```

# Preprocessing

```python
df = df_raw.copy()

# Force To Int
force_int_col = ['Age','Annual_Income','Monthly_Inhand_Salary','Num_of_Loan','Delay_from_due_date',
                 'Changed_Credit_Limit','Outstanding_Debt','Amount_invested_monthly','Monthly_Balance']

pattern = r'[\D]*'

for col in force_int_col:
    if df[col].dtypes not in ['O']: 
        continue

    try:
        df[col] = df[col].str.replace(pattern,'',regex=True).fillna(0).astype(int)
    except OverflowError:
        print(col,' too big, convert to float instead')
        df[col] = df[col].str.replace(pattern,'',regex=True).fillna(0).astype(float)
```

```python
dropped_col = []

for col in df.columns:
    print(col)
    dtypes = df[col].dtypes
    print(f'Dtypes',{dtypes})

    if dtypes == 'O':
        continue

    num_instances = df[col].nunique()
    print(f'Num Instance', num_instances)
    if num_instances <=10:
        print(f'Unique Instance',df[col].unique())

    print('Num Nulls',df[col].isnull().sum())
    print('-------')

    if dtypes == 'O' and num_instances >= 1000:
        dropped_col.append(col)
```
